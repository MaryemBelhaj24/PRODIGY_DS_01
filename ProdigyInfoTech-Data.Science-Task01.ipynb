```python
#Importing libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Reading Dataset
df=pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_426474.csv", skiprows=4)
df.head() 
```

```python
#Columns
df.columns
```

```python
# Remove columns with all NaN values
df = df.dropna(axis=1, how='all')
```

```python
df.head()
```

```python
#Data types of each column
df.dtypes
```

```python
#Shape
df.shape
```

```python
df.tail()
```

```python
#Information about the DataFrame
df.info()
```

```python
#Checking the count of null values 
df.isnull().sum()
```

```python
# Use interpolate to fill missing values in numerical columns
df=df.interpolate()
```

```python
#rechecking
df.isnull().sum()
```

```python
#Columns name change
df.columns=[each.split()[0] if(len(each.split())>2) else each.replace(" ","_") for each in df.columns]
df.head()
```

```python
#Unique values in the 'Country_Name' column
df.Country_Name.unique()
```

```python
#filter data
df2=df[['Country_Name', '2022']].iloc[:10]
df2
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x=df2['Country_Name'], y=df2['2022'], palette="magma")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Total Population by Country in 2022')
plt.yscale('log')  # Set the scale of the y-axis to be logarithmic
plt.xticks(rotation=90)
plt.show()
```

```python
# Sort the DataFrame based on the population in 2022 in descending order
df_sorted = df.sort_values(by='2022', ascending=False)
```

```python
# Select the top 10 countries
df_top10 = df_sorted.head(10)
```

```python
# Create the bar plot
plt.figure(figsize=(9,5))
sns.barplot(x=df_top10['Country_Name'], y=df_top10['2022'], palette="rocket")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries by Population in 2022')
plt.yscale('log')  # Set the scale of the y-axis to be logarithmic
plt.xticks(rotation=90)
plt.show()
```

```python
# Sort the DataFrame based on the population in 2022 in ascending order
df_sorted = df.sort_values(by='2022', ascending=True)

# Select the bottom 10 countries
df_bottom10 = df_sorted.head(10)

# Create the bar plot
plt.figure(figsize=(9,5))
sns.barplot(x=df_bottom10['Country_Name'], y=df_bottom10['2022'], palette="binary")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Bottom 10 Countries by Population in 2022')
plt.yscale('log')  # Set the scale of the y-axis to be logarithmic
plt.xticks(rotation=90)
plt.show()
```

```python
#Reading Dataset 2
df3=pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_426474.csv")
df3
```

```python
df3=df3.fillna("not classified")
df3
```

```python
df3.columns=[each.split()[0] if(len(each.split())>2) else each.replace(" ","_") for each in df3.columns]
# Merge the two DataFrames on the 'country code' column
merged_df = pd.merge(df, df3, on='Country_Code')
merged_df
```

```python
#Creating a 2x2 subplot grid
f, axs = plt.subplots(2, 2, figsize=(10, 10), constrained_layout=True)
# Filter data for each income group
high_income_df = merged_df[merged_df['IncomeGroup'] == 'High income']
low_income_df = merged_df[merged_df['IncomeGroup'] == 'Low income']
upper_middle_income_df = merged_df[merged_df['IncomeGroup'] == 'Upper middle income']
lower_middle_income_df = merged_df[merged_df['IncomeGroup'] == 'Lower middle income']

# Sort and select top 10 countries for each income group
sort_high = high_income_df.sort_values(by=['2022', '2021', '2020'], ascending=False).head(10)
sort_low = low_income_df.sort_values(by=['2022', '2021', '2020'], ascending=False).head(10)
sort_upper_middle = upper_middle_income_df.sort_values(by=['2022', '2021', '2020'], ascending=False).head(10)
sort_lower_middle = lower_middle_income_df.sort_values(by=['2022', '2021', '2020'], ascending=False).head(10)

# Reshape the data to "long" format
sort_high_long = pd.melt(sort_high, id_vars='Country_Name', value_vars=['2022', '2021', '2020'], var_name='Year', value_name='High income')
sort_low_long=pd.melt(sort_low,id_vars='Country_Name',value_vars=['2022', '2021', '2020'], var_name='Year', value_name='low income')
sort_upper_middle_long=pd.melt(sort_upper_middle,id_vars='Country_Name',value_vars=['2022', '2021', '2020'], var_name='Year', value_name='upper middle income')
sort_lower_middle_long=pd.melt(sort_lower_middle,id_vars='Country_Name',value_vars=['2022', '2021', '2020'], var_name='Year', value_name='lower middle income')

# Create the bar plot 
ax = sns.barplot(x="Country_Name", y="High income", hue="Year", data=sort_high_long, ax=axs[0, 0], palette="viridis")
ax.set_yscale("log")
ax.tick_params(axis='x', rotation=45)
ax.set_title('Top 10 High Income Countries Over Years')

ax = sns.barplot(x="Country_Name", y="low income", hue="Year", data=sort_low_long, ax=axs[0, 1], palette="viridis")
ax.set_yscale("log")
ax.tick_params(axis='x', rotation=45)
ax.set_title('Top 10 low Income Countries Over Years')


ax = sns.barplot(x="Country_Name", y="upper middle income", hue="Year", data=sort_upper_middle_long, ax=axs[1, 0], palette="viridis")
ax.set_yscale("log")
ax.tick_params(axis='x', rotation=45)
ax.set_title('Top 10 upper middle Income Countries Over Years')


ax = sns.barplot(x="Country_Name", y="lower middle income", hue="Year", data=sort_lower_middle_long, ax=axs[1, 1], palette="viridis")
ax.set_yscale("log")
ax.tick_params(axis='x', rotation=45)
ax.set_title('Top 10 lower middle Income Countries Over Years')

#Displaying the subplots
plt.show()
```
